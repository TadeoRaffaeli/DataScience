  <h2 style="color: #d35400;">Variables de nuestro Data Set 📝</h2>
  <ol style="color: #333;">
  <li><strong>fixed acidity:</strong> la mayoría de los ácidos involucrados con el vino o fijos no volátiles (no se evaporan fácilmente).</li>
  <li><strong>volatile acidity:</strong> la cantidad de ácido acético en el vino, que en niveles demasiado altos puede provocar un sabor desagradable a vinagre.</li>
  <li><strong>citric acid:</strong> encontrado en pequeñas cantidades, el ácido cítrico puede agregar 'frescura' y sabor a los vinos.</li>
  <li><strong>residual sugar:</strong> cantidad de azúcar que queda en el vino posterior a la fermentación del mismo. Se considera raro encontrar vinos con menos de 1 gramo por litro. Los vinos con 45 gramos por litro se consideran vinos dulces.</li>
  <li><strong>chlorides:</strong> cantidad de sal en el vino.</li>
  <li><strong>free sulfur dioxide:</strong> la forma libre de SO2 existe en equilibrio entre el SO2 molecular (como gas disuelto) y el ión bisulfito; previene el crecimiento microbiano y la oxidación del vino.</li>
  <li><strong>total sulfur dioxide:</strong> cantidad de formas libres y ligadas de S02; en bajas concentraciones, el SO2 es mayormente indetectable en el vino, pero en concentraciones de SO2 libres superiores a 50 ppm, el SO2 se vuelve evidente en la nariz y el sabor del vino.</li>
  <li><strong>density:</strong> la densidad del agua es cercana a la del agua dependiendo del porcentaje de contenido de alcohol y azúcar.</li>
  <li><strong>pH:</strong> describe qué tan ácido o básico es un vino en una escala de 0 (muy ácido) a 14 (muy básico); la mayoría de los vinos están entre 3 y 4 en la escala de pH.</li>
  <li><strong>sulphates:</strong> un aditivo del vino que puede contribuir a los niveles de dióxido de azufre (S02), que actúa como antimicrobiano y antioxidante.</li>
  <li><strong>alcohol:</strong> el porcentaje de contenido de alcohol del vino.</li>
  <li><strong>quality:</strong> puntaje del 0 al 10, 10 siendo el más alto. Se considera a un vino de 6.5 un vino "Bueno".</li>
  </ol>

En el primer gráfico de radar, dividimos los valores de pH en 10 intervalos para que todas las variables estuvieran en el mismo rango (entre 0 y 1) y fueran más fáciles de visualizar. Esto nos permitió analizar visualmente cómo las variables, como el pH, afectan las distintas calidades de vino.</p>

```python
# Agrupar los datos por calidad y calcular las medias
df_quality = df.groupby('quality').mean()

# Convertir el índice 'quality' en una columna
df_quality.reset_index(inplace=True)

# Variables seleccionadas
variables = ['citric acid', 'volatile acidity', 'chlorides', 'sulphates', 'pH']

# Escalar los valores de la variable 'pH' dividiéndolos por 10, para tener valores proporcionales a los otros analisados.
df_quality['pH'] = df_quality['pH'] / 10
```

```python
grupos = ['Calidad 5', 'Calidad 6', 'Calidad 7', 'Calidad 8']
datos_por_grupo = [df_quality[df_quality['quality'] == 5][variables].values[0],
                   df_quality[df_quality['quality'] == 6][variables].values[0],
                   df_quality[df_quality['quality'] == 7][variables].values[0],
                   df_quality[df_quality['quality'] == 8][variables].values[0]]

# Preparar el gráfico de radar
angulos = 2 * np.pi * np.linspace(0, 1, len(variables), endpoint=False)
angulos = np.concatenate((angulos, [angulos[0]]))  # Repetir el primer ángulo para cerrar el polígono

# Crear el gráfico de radar para cada grupo
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 8), subplot_kw=dict(polar=True))

for i, ax in enumerate(axs.flat):
    datos = datos_por_grupo[i]
    datos = np.concatenate((datos, [datos[0]]))  # Repetir el primer valor para cerrar el polígono

    ax.plot(angulos, datos, linewidth=2, label=grupos[i])
    ax.fill(angulos, datos, alpha=0.25)
    ax.set_xticks(angulos[:-1])
    ax.set_xticklabels(variables)
    ax.set_title(grupos[i])
    ax.legend()

plt.suptitle('Gráficos de Radar', fontsize=16)
plt.tight_layout()
plt.show()
```

  <h2 style="color: #d35400;">Limpieza de Outliers</h2>
  <p style="color: #333;">A continuación, llevamos a cabo una limpieza de los valores atípicos en nuestro conjunto de datos. Inicialmente, consideramos el enfoque tradicional de eliminar los valores que se encontraban fuera del rango de ±2 veces el rango intercuartil (IQR). Sin embargo, notamos que esto eliminaba un número significativamente mayor de valores de lo esperado.</p>
  
  <p style="color: #333;">Para abordar este problema y evitar la eliminación excesiva de datos, decidimos ajustar nuestros criterios y utilizar un rango más amplio, ±2 veces el IQR, para definir los valores atípicos. Esto nos brinda un margen adicional y permite conservar más datos, especialmente aquellos relacionados con vinos de mejor calidad, que a menudo tienen niveles más altos de sulfitos. Esta decisión se tomó para evitar sesgar nuestro análisis hacia las categorías de menor calidad y mantener una muestra más representativa de vinos de calidad superior.</p>
</div>

```python
for variable in ['volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']:
    # Calcular el rango intercuartílico (IQR)
    Q1 = df[variable].quantile(0.25)
    Q3 = df[variable].quantile(0.75)
    IQR = Q3 - Q1

    # Definir los límites para identificar outliers
    lower_limit = Q1 - 2 * IQR
    upper_limit = Q3 + 2 * IQR

    # Filtrar las filas que no contienen outliers
    df = df[(df[variable] >= lower_limit) & (df[variable] <= upper_limit)]
```

<h2 style="color: #d35400;">Eliminación de Outliers🧠: </h2>
<p style="color: #333;">Para mejorar la calidad del análisis de datos, se creó una función llamada eliminar_outliers. Esta función permite eliminar valores atípicos (outliers) en un DataFrame, lo que es esencial para garantizar que los resultados no se vean afectados por valores extremos. La función incluye una opción para excluir una variable específica, lo que puede ser útil si se sospecha que una variable en particular tiene una gran cantidad de valores atípicos.</p>

<p style="color: #333;">Se aplicó la función eliminar_outliers en ambos DataFrames, primero excluyendo la variable "fixed acidity" y luego sin exclusiones. Esto se realizó ya que Se ha determinado que esta variable tiene una correlación relativamente baja con otras variables del conjunto de datos, especialmente en comparación con las fuertes correlaciones observadas entre otras variables.</p>









```python
# Definir la función eliminar_outliers con opción para excluir variable
def eliminar_outliers(df, exclude_variable=None):
    df_outliers = df.copy()
    variables_numericas = ['volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

    if exclude_variable:
        variables_numericas.remove(exclude_variable)

    for variable in variables_numericas:
        Q1 = df_outliers[variable].quantile(0.25)
        Q3 = df_outliers[variable].quantile(0.75)
        IQR = Q3 - Q1

        lower_limit = Q1 - 2 * IQR
        upper_limit = Q3 + 2 * IQR

        df_outliers = df_outliers[(df_outliers[variable] >= lower_limit) & (df_outliers[variable] <= upper_limit)]

    return df_outliers

# Cargar los DataFrames desde los archivos CSV
df_mala = pd.read_csv('data_calidad_Mala.csv')
df_buena = pd.read_csv('data_calidad_Buena.csv')

# Eliminar outliers en cada DataFrame
df_mala = eliminar_outliers(df_mala)
df_buena = eliminar_outliers(df_buena)

print("\nDataFrame Mala:")
print(df_mala.shape)

print("\nDataFrame Buena :")
print(df_buena.shape)
```

  <h2 style="color: #333;">SMOTE</h2>
  <p style="color: #333;">SMOTE (Synthetic Minority Over-sampling Technique) es una técnica de sobre muestreo que se utiliza comúnmente en conjuntos de datos desbalanceados. En problemas de clasificación, los conjuntos de datos desbalanceados tienen una distribución irregular entre las clases, donde una clase (la clase minoritaria) tiene significativamente menos ejemplos que otra (la clase mayoritaria).</p>
  <p style="color: #333;">Si bien nuestra base de datos tiene un balanceo bastante parejo luego de eliminar duplicados y Outliers, vamos a ver qué pasaría si aplicamos SMOTE y tenemos dos categorías perfectamente balanceadas.</p>

</div>

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

X = df_unificado.drop('quality', axis=1)  # Características
y = df_unificado['quality']  # Etiquetas

print("Antes de aplicar SMOTE:", Counter(y))
```

```python
smote = SMOTE(sampling_strategy='auto')
X_resampled, y_resampled = smote.fit_resample(X, y)

print("Después de aplicar SMOTE:", Counter(y_resampled))
```

  <h2 style="color: #8B4513;">Stratified K-Fold en Data Science</h2>
  <p style="color: #333;">Stratified K-Fold es una técnica de validación cruzada utilizada en Machine Learning para evaluar modelos. Esta técnica es especialmente útil cuando se trabaja con conjuntos de datos desequilibrados en cuanto a la distribución de las clases. En lugar de dividir aleatoriamente el conjunto de datos en diferentes folds, Stratified K-Fold realiza la división manteniendo la proporción de clases en cada fold.</p>
  <p style="color: #333;">En nuestro análisis, Stratified K-Fold podría ser crucial para evaluar la capacidad de generalización de nuestro modelo de Machine Learning.</p>
  <p style="color: #333;">Al utilizar Stratified K-Fold, nos aseguramos que cada fold tenga una distribución de clases similar a la del conjunto de datos original. Esto ayuda a obtener estimaciones más precisas del rendimiento del modelo en datos no vistos, lo que puede ser fundamental para la toma de decisiones.</p>
</div>

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
# Separar en X y y
data = df_unificado.values

X, y = data[:, :-1], data[:, -1]
print(X.shape, y.shape)
```

```python
# Definir el modelo de Random Forest con los hiperparámetros
best_rf_model = RandomForestClassifier(
    max_depth=None,
    max_features='sqrt',
    min_samples_leaf=1,
    min_samples_split=10,
    n_estimators=100,
    random_state=42
)

# Crear el objeto Stratified K-Fold con 5 splits
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Listas para almacenar las métricas
accuracy_scores = []
precision_scores = []

# Ciclo para iterar sobre los diferentes folds
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Entrenar el modelo
    best_rf_model.fit(X_train, y_train)

    # Realizar predicciones
    y_pred = best_rf_model.predict(X_test)

    # Calcular métricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)

    # Agregar las métricas a las listas
    accuracy_scores.append(accuracy)
    precision_scores.append(precision)

# Calcular el promedio de las métricas
avg_accuracy = sum(accuracy_scores) / len(accuracy_scores)
avg_precision = sum(precision_scores) / len(precision_scores)

# Imprimir los resultados promedio
print(f"Accuracy promedio usando Stratified K-Fold: {avg_accuracy:.4f}")
print(f"Precisión promedio usando Stratified K-Fold: {avg_precision:.4f}")
```
